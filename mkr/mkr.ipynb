```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.linear_model import LinearRegression

df1 = pd.DataFrame({
    "X1": [2, 5, 1, 2, 1, 1],
    "X2": [529000, 360916, 840000, 1699000, 620000, 865000],
    "X3": [3, 2, 2, 2, 2, 2],
    "X4": [53.84, 39.23, 47,65.55, 43, 51.59],
    "X5": [2021, 2022, 2019, 2009, 1990, 2020]
})

df2 = pd.DataFrame({
    "Y":  [529, 361, 840, 1699, 620],
    "X1": [3, 2, 2, 2, 2],
    "X2": [53, 39, 47, 65, 43],
    "X3": [2021, 2022, 2019, 2009, 1990]
})

print(df1)
print("===============================")
print(df2)
```

<b> Завдання 1 </b> <br> 
Для даних наведених в таблиці зробити ієрархічну кластеризацію та намалювати дендограму. Для кластеризації використовувати евклідову метрику.

```python
lm = linkage(df1, method='ward', metric="euclidean")

dendrogram(lm) 
plt.title("Дендрограма") 
plt.show() 
```

<b> Завдання 2 </b> <br> 
Побудувати множинну регресію для даних визначених нижче. У відповіді написати рівняння регресії та R².

```python
X = df2[["X1", "X2", "X3"]]
y = df2["Y"]

model = LinearRegression()
model.fit(X, y)

coef = model.coef_
intercept = model.intercept_
r2 = model.score(X, y)

print("=== Рівняння регресії ===")
print(f"Y = {intercept:.5f} + ({coef[0]:.5f})*X1 + ({coef[1]:.5f})*X2 + ({coef[2]:.5f})*X3")
print("\n=== R² моделі ===")
print(f"R² = {r2:.7f}")
```
